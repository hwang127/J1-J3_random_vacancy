```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 152
Y=[[np.sqrt(3) * i / 2]* L for i in range(L)]
X=[np.arange(L) -0.5 * i for i in range(L)]
```

```python
eyy = np.loadtxt(f"eyy_{L}.txt")
exx = -eyy
exy = np.loadtxt(f"exy_{L}.txt")
```

```python
def strain(i0, j0, L):
    new_exx, new_eyy, new_exy = np.zeros((L, L)), np.zeros((L, L)), np.zeros((L, L))
    for i in range(L):
        for j in range(L):
            di = i - i0
            dj = j - j0
            new_i = (L//2 + di) % L
            new_j = (L//2 + dj) % L
            new_exx[i, j] = exx[new_i, new_j]
            new_eyy[i, j] = eyy[new_i, new_j]
            new_exy[i, j] = exy[new_i, new_j]
    return new_exx, new_eyy, new_exy

def sample(num):
    xx, yy, xy = np.zeros((L, L)), np.zeros((L, L)), np.zeros((L, L))
    n = L**2 // num
    #n = num
    locations = np.random.randint(0, L, size = (n, 2))
    for tup in locations:
        new_exx, new_eyy, new_exy = strain(tup[0], tup[1], L)
        xx += new_exx
        yy += new_eyy
        xy += new_exy
    return np.concatenate((xx, yy, xy), axis = 0)
```

```python
num = 100
for i in range(11, 101):
    D = sample(num)
    print(D.shape)

    np.savetxt(f'F:\\CM\\Prjt2\\Simulation\\newUpdate\\vacancy\\{L}\\{i}_{L}_{num}_sample.txt', D, )
```

```python
ret = sample(10000)
```
